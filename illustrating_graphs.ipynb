```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

```

```python
# Create a graph
G = nx.Graph()


nodes = [i for i in range(65)]
# Add nodes
for node in nodes:
    G.add_node(node)

res_nodes = []
for node in nodes:
    if node == 0:
        continue
    G.add_edge(node, node-1)
    if node % 9 == 0:
        for res_node in res_nodes:
            if res_node < 44:
                G.add_edge(res_node, node)
        res_nodes.append(node)
    if node == len(nodes):
        G.add_edge(node, 1)

print(G.nodes)

```

```python

node_colors = ['blue' if node < 44 else "green" for node in nodes ]

print(G)
# Draw the graph in a circular layout
pos = nx.circular_layout(G)
plt.figure(figsize=(10, 10))

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_color=node_colors)

# Draw edges
nx.draw_networkx_edges(G, pos)

blue_patch = mpatches.Patch(color='blue', label='ingredient nodes')
green_patch = mpatches.Patch(color='green', label='compound nodes')
plt.legend(handles=[blue_patch, green_patch])
# Display the graph
plt.axis("off")
plt.show()
```
