```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 100)

y1 = 0.2 * x + 0.4
y1 = y1 ** 2

y2 = 0.6 - 0.2 * x
y2 = y2 ** 2

y3 = np.where(x < 0.5, 0.2 * x + 0.45, 0.65 - 0.2 * x)
y3 = y3 ** 2

y4 = 1 - y3
y4 = (y4 ** 2) - 0.333

fig, axs = plt.subplots(2, 2, figsize=(6, 6))

axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Substitutability Boosted')
axs[0, 0].set_xlim(0, 1)
axs[0, 0].set_ylim(0, 1)
axs[0, 0].set_xticks([0, 1])
axs[0, 0].set_yticks([0, 1])
axs[0, 0].set_xlabel('Substitutability')
axs[0, 0].set_ylabel('Negative Sample Likelyhood')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Non-substitutability Boosted')
axs[0, 1].set_xlim(0, 1)
axs[0, 1].set_ylim(0, 1)
axs[0, 1].set_xticks([0, 1])
axs[0, 1].set_yticks([0, 1])
axs[0, 1].set_xlabel('Substitutability')
axs[0, 1].set_ylabel('Negative Sample Likelyhood')

axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Center Boosted')
axs[1, 0].set_xlim(0, 1)
axs[1, 0].set_ylim(0, 1)
axs[1, 0].set_xticks([0, 1])
axs[1, 0].set_yticks([0, 1])
axs[1, 0].set_xlabel('Substitutability')
axs[1, 0].set_ylabel('Negative Sample Likelyhood')

axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Fringes Boosted')
axs[1, 1].set_xlim(0, 1)
axs[1, 1].set_ylim(0, 1)
axs[1, 1].set_xticks([0, 1])
axs[1, 1].set_yticks([0, 1])
axs[1, 1].set_xlabel('Substitutability')
axs[1, 1].set_ylabel('Negative Sample Likelyhood')

# Adjusting layout
plt.tight_layout()
plt.show()
```
